```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("./images/bird.jpg")
```

```python
def display_image(plotName, image, cmap = None):
    plt.figure(figsize=(10,5))
    plt.subplot(1,1,1)
    plt.imshow(image, cmap=cmap)
    plt.title(plotName)
    
```

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
display_image("Bird", img_rgb)
```

```python
## access first 100 rows
corners = img_rgb[0:100, 0:100]
```

```python
display_image('corners',corners)
```

### Changing colors

```python
green = (0,255,0)
img_rgb[0:100, 0:100] = green
```

```python
display_image("Manipulated green", img_rgb)
```
