## Convert Color

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
## carga por defecto en formato bgr (razones historicas desde los 90s)

img = cv.imread('./images/flemingo.jpg')
```

```python
img
```

```python
## 3ra dimension son los canales de color, en este caso 3 bgr
img.shape
```

### Split

```python
b,g,r = cv.split(img)
```

```python
b
```

## Color Convertor
cv2 usa por defecto bgr por razones historicas desde los 90.
Matplot lib funciona con rgb asi que hay que convertirlo

```python
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,4,1)
plt.imshow(img_rgb)
plt.title('color')

plt.subplot(1,4,2)
plt.imshow(b,cmap='gray')
plt.title('blue')

plt.subplot(1,4,3)
plt.imshow(r, cmap = 'gray')
plt.title('red')

plt.subplot(1,4,4)
plt.imshow(g, cmap = 'gray')
plt.title('green')

plt.show()
```

```python
plt.imshow(b)# SIN cmap → colores raros, mezcla viridis
```
