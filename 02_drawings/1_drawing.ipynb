```python
import numpy as np
import cv2 as cv

import sys, os
sys.path.insert(0, os.path.abspath(".."))
from helpers import display_image
```

```python
canvas = np.zeros((300,300,3), dtype="uint8")
```

```python
display_image('canvas',canvas)
```

```python
green = (0,255,0)
cv.line(canvas, (0,0), (300, 300), green)
display_image('canvas',canvas)
```

```python
red = (255,0,0)
cv.line(canvas, (300,0), (0, 300), red, 3)
display_image("canvas red line", canvas)
```

```python
#drawing rectangle
cv.rectangle(canvas, (10,10), (60,60),red,3 )
display_image('canvas',canvas)
```

```python
cv.rectangle(canvas, (50,200), (200,255), green, -1)
display_image('canvas', canvas)
```

## Circles

```python
canvas_circle = np.zeros((300,300,3), dtype="uint8")
```

```python
center_x, center_y = canvas.shape[1] // 2, canvas.shape[0] // 2
white = (255,255,255)

for r in range(0,175,25):
    cv.circle(canvas_circle, (center_x, center_y), r, white)
    
display_image('circles',canvas_circle)
```

## abstract drawing

```python
# just playing around

for i in range(0,25):
    radius = np.random.randint(5,200)
    color = np.random.randint(0,256, size=(3,)).tolist()
    pt=np.random.randint(0,300,size=(2,))
    
    cv.circle(canvas,tuple(pt),radius,color,-1)
    
display_image('canvas', canvas)
```
